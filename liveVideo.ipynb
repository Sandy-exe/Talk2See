```python
def is_blurry(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    fm = cv2.Laplacian(gray, cv2.CV_64F).var()
    return fm < 1000,fm  # Adjust this threshold as needed
```

```python
import io
import threading
import time

def speak(frame,text):

    if text == "Place the camera Properly":
        engine.say(text)
        engine.runAndWait()
        return
    
    chatgpt_api_Ocr = chatgpt_api(get_OCR(frame))
    if chatgpt_api_Ocr == "ungomma69":
        engine.say("Sorry I couldn't read the text, Trying again...")
        engine.runAndWait()
        return
    engine.say(chatgpt_api_Ocr)
    engine.runAndWait()
```

```python
import threading

# Flags to check if a thread is already running
speak_thread_running = False
ocr_thread_running = False

cap = cv2.VideoCapture(2)
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
  Blurry = False
  start_time = time.time()
  while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        print("Ignoring empty camera frame.")
        break

    image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    image.flags.writeable = False
    results = pose.process(image)  # make detections
    image.flags.writeable = True
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

    # extract landmarks
    try:
       
       blurry_score,fm = is_blurry(frame)

       cv2.putText(image, str(round(fm,2)), (10, 100),
                   cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)
       
       if time.time() - start_time > 5 and Blurry:
            start_time = time.time()
            Blurry = False
       
       if blurry_score and not Blurry:
          Blurry = True
          if not speak_thread_running:
              speak_thread = threading.Thread(target=speak, args=(frame,"Place the camera Properly",))
              speak_thread.start()
              speak_thread_running = True
          print("Blurry")

       if Blurry and not blurry_score:
            Blurry = False
            if not ocr_thread_running:   
                ocr_thread = threading.Thread(target=speak, args=(frame,"not blurry"))
                ocr_thread.start()
                ocr_thread_running = True
            print("Not Blurry")
            
    except:
       
       pass

    # Reset the flags if the threads have finished running
    if speak_thread_running and not speak_thread.is_alive():
        speak_thread_running = False
    if ocr_thread_running and not ocr_thread.is_alive():
        ocr_thread_running = False

    # draw the thing
    cv2.imshow('Mediapipe Feed', image)
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
    
cap.release()
cv2.destroyAllWindows()
```
